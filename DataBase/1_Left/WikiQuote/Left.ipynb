Naom Chomsky

```python
def left_quotes(author, file_name):#, nationality):
    quotes = wikiquotes.get_quotes(author, 'english')
    csv = '.csv'
    df = pd.DataFrame({'Quotes': quotes,
                       'Author': author,
                       'Political Orientation': 'Left'})#,
#                       'Nationality': nationality})
    df.to_csv(file_name+csv, index=False)#, sep='\t',encoding='utf-8')
    return df
```

```python
Chomsky = left_quotes('Noam Chomsky', 'Chomsky')
```

```python
Chomsky_Reader = left_quotes('The Chomsky Reader', 'Chomsky_Reader')
```
